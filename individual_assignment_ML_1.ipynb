
**Encoding adalah** proses mengubah data kategorikal menjadi bentuk yang dapat dimengerti oleh model atau algoritma pembelajaran mesin. Pada dasarnya, ini melibatkan mengonversi data kategorikal (seperti teks atau kategori) menjadi representasi numerik yang dapat digunakan oleh model. Dua metode encoding umum adalah Label Encoding dan One-Hot Encoding.

**Label Encoding:**

Pada Label Encoding, setiap kategori diberikan suatu label unik dalam bentuk bilangan bulat.

Misalnya, jika kita memiliki kolom "Gender" dengan nilai "Male" dan "Female", Label Encoding dapat menggantikan "Male" dengan 0 dan "Female" dengan 1.

**Keuntungan:**

Efisien untuk kolom dengan banyak nilai kategori.
Kerugian:
**bold text**
Model mungkin mengartikan bahwa nilai yang lebih tinggi memiliki tingkat yang lebih tinggi, yang bisa menjadi masalah jika tidak ada hubungan ordinal yang sebenarnya.

**Encoding yang dipakai:**
-> one hot encoding

alasan: One-Hot Encoding dipilih karena kategori-kategori dalam dataset, seperti 'room_type' dan 'neighbourhood_group', tidak memiliki urutan intrinsik atau tingkatan. Metode ini mendukung penanganan kategori multinomial dan mencegah model dari membuat asumsi yang tidak benar tentang hubungan ordinal antar kategori. Dengan One-Hot Encoding, setiap kategori direpresentasikan oleh kolom terpisah, memastikan independensi antar kategori dan menghindari risiko overfitting, terutama ketika terdapat banyak kategori. Pilihan antara One-Hot Encoding dan Label Encoding disesuaikan dengan sifat data dan kebutuhan model.

**Normalisasi adalah** proses mengubah nilai-nilai variabel menjadi rentang yang seragam atau umum, sehingga perbedaan skala antar variabel tidak mempengaruhi kinerja model atau analisis data. Tujuan normalisasi adalah untuk membawa variabel-variabel ke dalam skala yang serupa, sehingga model atau algoritma pembelajaran mesin dapat menginterpretasikan dan memprosesnya dengan lebih baik.

**Normalisasi umumnya** dilakukan pada variabel numerik. Dua metode normalisasi yang umum digunakan adalah Min-Max Scaling dan Z-score Standardization.

Dalam beberapa contoh normalisasi yang umum, kita dapat menggunakan Min-Max Scaling untuk mengubah variabel numerik menjadi rentang 0 hingga 1.seperti contohnya menggunakan Min-Max Scaling pada beberapa kolom numerik seperti 'latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365'

**alasan menggunakan min max scaling:**

1. **Rentang yang Seragam:**
   - Metode ini memastikan bahwa nilai-nilai dalam setiap kolom numerik berada dalam rentang yang seragam, biasanya antara 0 dan 1. Hal ini dapat membantu menghilangkan perbedaan skala antar variabel, sehingga tidak ada variabel yang mendominasi yang lain dalam analisis.

2. **Penting untuk Algoritma Berbasis Jarak:**
   - Algoritma yang menggunakan perhitungan jarak, seperti k-Nearest Neighbors atau K-Means, dapat dipengaruhi oleh perbedaan skala antar variabel. Normalisasi Min-Max membantu mencegah variabel dengan skala besar mendominasi perhitungan jarak.

3. **Interpretasi yang Lebih Mudah:**
   - Nilai-nilai yang ternormalisasi antara 0 dan 1 lebih mudah diinterpretasikan, terutama jika hasil tersebut perlu dijelaskan atau disampaikan kepada pemangku kepentingan yang tidak familiar dengan skala asli variabel.

4. **Tidak Sensitif terhadap Outliers:**
   - Min-Max Scaling kurang sensitif terhadap pencilan (outliers) dibandingkan beberapa metode normalisasi lainnya. Ini karena rentang skala diukur dari nilai minimum dan maksimum, yang tidak terlalu dipengaruhi oleh pencilan.
